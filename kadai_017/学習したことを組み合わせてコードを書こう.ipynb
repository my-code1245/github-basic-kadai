```python
class Human:
 def __init__(self,name,age):
  self.name = name
  self.age = age

 def check_adult(self):
  if self.age >=20:
   print(f"{self.name}は成人です")

  else:
    print(f"{self.name}は未成年です")

 humans = [Human("侍太郎",25),
          Human("侍花子",19),
          Human("侍ぷー太郎",30) ]

 for human in humans:
  human.check_adult()








```
